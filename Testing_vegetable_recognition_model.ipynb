```python
test_set = tf.keras.utils.image_dataset_from_directory (
    '/content/drive/MyDrive/vegetable_fruits_recognition/test',
    labels='inferred',
    label_mode='categorical',
    class_names=None,
    color_mode='rgb',
    batch_size=32,
    image_size=(64,64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation='bilinear',
    follow_links=False,
    crop_to_aspect_ratio=False,
    )
```

#loading model

```python
cnn = tf.keras.models.load_model('/content/trained_model.h5')
```

##visualization performin prediction on single image

```python
import cv2
image_path  = '/content/drive/MyDrive/vegetable_fruits_recognition/test/eggplant/Image_1.jpg'
img = cv2.imread(image_path)
plt.imshow(img)
plt.title('test_image')
plt.xticks([])
plt.yticks([])
plt.show()
```

#testing model

```python
image= tf.keras.preprocessing.image.load_img(image_path,target_size=(64,64))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])   #converting single image to batch
prediction = cnn.predict(input_arr)
print(prediction)
```

```python
test_set.class_names
```

```python
result_index = np.where(prediction[0] == max(prediction[0]))
print(result_index)
```

```python
#display image
plt.imshow(img)
plt.title('test_image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#perform single prediction
print("its a {}" .format(test_set.class_names[result_index[0][0]]))
```
